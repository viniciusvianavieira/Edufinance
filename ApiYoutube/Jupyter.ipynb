```python
import json
import pandas as pd

file ="brenno_sullivan.json"
dados = None
with open(file,'r') as f:
    dados = json.load(f)

id_canal, estatistica = dados.popitem()
print(id_canal)
estatistica_canal = estatistica["channel_statistics"]
estatistica_video = estatistica["video_data"]
```

```python
#estatisticas canal
visualicacoes = estatistica_canal["viewCount"]
inscricoes = estatistica_canal["subscriberCount"]
qtdvideos =  estatistica_canal["videoCount"]

print("Visualizações:",visualicacoes)
print('Inscrição:',inscricoes )
print('Quantidade videos:',qtdvideos )

```

```python
#estatisticas videos

sorted_vids = sorted(estatistica_video.items(),key=lambda item: int(item[1]["viewCount"]), reverse = True)
estatistica = []
for vid in sorted_vids:
    id_videos = vid[0]
    titulo = vid[1]["title"]
    visualicacoes = vid[1]["viewCount"]
    likes = vid[1]["likeCount"]
    #faltou um dislike????
    comentarios = vid[1]["commentCount"]
    estatistica.append([titulo, visualicacoes, likes,comentarios])
```
